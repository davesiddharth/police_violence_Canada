```python
import pandas as pd
```

```python
file_path = "fatal_police_encounter_2000_2020.csv"
```

```python
fatal_police_df = pd.read_csv(file_path)
```

```python
fatal_police_df.head()
```

```python
fatal_police_dropc_df = fatal_police_df.drop(columns=['NAME MIDDLE','NICKNAME','PICTURE SOURCE'])
```

```python
fatal_police_dropc_df.head(50)
```

```python
fatal_police_dropc_df[['YEAR','MONTH','DATE']]=fatal_police_dropc_df.DATE.str.split("-",expand=True,)
```

```python
fatal_police_encounter_df = fatal_police_dropc_df[['VICTIM ID','DATE','MONTH','YEAR','NAME FIRST','NAME LAST','AGE','GENDER','RACE','PROV','POLICE SERVICE','CAUSE DEATH','ARMED TYPE','SUMMARY']]
```

```python
fatal_police_encounter_df.head()
```

```python
fatal_police_encounter_df.isnull().sum()
```

```python
fatal_police_encounter_df.fillna("Unknown", inplace=True)
```
